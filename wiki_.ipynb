```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_Indian_films'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
achi_table = soup.find_all('table')[1]
```

```python
print(achi_table)

```

```python
titles = achi_table.find_all('th')
print(titles)
# th tags is for titles
```

```python
table_titles = [title.text.strip() for title in titles]

# list comprehension
```

```python
print(table_titles)
```

```python
# get rid of \n
# The .strip() method is a built-in string method in Python that is used to remove leading and trailing whitespace characters (spaces, tabs, newlines, etc.)
#  from a string. It does not remove whitespace within the string, only from the beginning and end.
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)
df
```

```python
# td tags is for the data rows
# tr represnets the rows, td represents the data within that row
```

```python
column_data = achi_table.find_all('tr')
```

```python
for row in column_data[1:]:
    td_elements = row.find_all('td')
    individual_row_data = [data.text.strip() for data in td_elements]
    # print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

So, df.loc[length] is a way to access a specific row in the DataFrame based on its index and then assign values to that row. It's commonly used for adding new rows to a DataFrame, especially when you want to dynamically append data to an existing DataFrame
